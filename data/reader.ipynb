```python
import numpy as np
import os

# 定义动作映射表
action_map = {
    'A41': 'sneeze/cough',
    'A42': 'staggering',
    'A43': 'falling down',
    'A44': 'headache',
    'A45': 'chest pain',
    'A46': 'back pain',
    'A47': 'neck pain',
    'A48': 'nausea/vomiting',
    'A49': 'fan self',
    'A103': 'yawn',
    'A104': 'stretch oneself',
    'A105': 'blow nose'
}

def read_skeleton(file_path):
    # 读取骨骼文件
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # 假设每个文件的第一行是帧数
    frame_count = int(lines[0].strip())

    # 初始化存储数组，这里需要根据实际的关节数和每个关节的数据点数调整
    # 假设有25个关节，每个关节有3个坐标点 (x, y, z)
    skeleton_data = np.zeros((frame_count, 25, 3), dtype=float)

    # 解析文件中的数据
    line_idx = 1  # 从第二行开始
    for f in range(frame_count):
        joint_count = int(lines[line_idx].strip())  # 假设每个帧的第一行是关节数
        line_idx += 1
        for j in range(joint_count):
            joint_data = lines[line_idx].strip().split()
            # 假设关节数据按照 x, y, z 的顺序排列
            skeleton_data[f, j, :] = [float(joint_data[0]), float(joint_data[1]), float(joint_data[2])]
            line_idx += 1

    return skeleton_data

def load_data(folder_path):
    # 存储所有动作的数据
    actions_data = []

    for file_name in os.listdir(folder_path):
        if file_name.endswith('.skeleton'):
            action_id = file_name[17:21]  # 提取文件名中的动作ID
            if action_id in action_map:
                file_path = os.path.join(folder_path, file_name)
                skeleton_data = read_skeleton(file_path)
                actions_data.append(skeleton_data)

    return np.array(actions_data)

# 示例：加载数据
folder_path = 'C:/Users/zhaor/Desktop/nturgb+d_skeletons'
# actions_data = load_data(folder_path)

# 打印动作数据的维度
# print(actions_data.shape)
```
